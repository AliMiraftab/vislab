# Wikipaintings

This dataset of ~100K paintings and associated genre and style labels (and free-form tags) is scraped from http://wikipaintings.org.

```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import vislab.dataset
import vislab.datasets
import vislab.dataset_viz
import vislab.dataset_stats
import re
```

Load basic dataset, which is scraped entirely from the search results page and includes the page urls of the artworks.

```python
basic_df = vislab.datasets.wikipaintings.get_basic_df()
basic_df.iloc[:5]
```

Load detailed dataset, which is scraped from the individual artwork pages and includes the labels and image urls of the artworks.

```python
df = vislab.datasets.wikipaintings.get_df()
print df.shape
df.iloc[:5]
```

Expand the style label into own DataFrame.

```python
df_style = vislab.datasets.wikipaintings.get_style_df(1000)
```

### Style, Genre, Technique, and Artist distribution plots.

```python
fig = vislab.dataset_viz.plot_column_frequencies(df, 'style', 30)
fig = vislab.dataset_viz.plot_column_frequencies(df, 'genre', 20)
fig = vislab.dataset_viz.plot_column_frequencies(df, 'technique', 20)
fig = vislab.dataset_viz.plot_column_frequencies(df, 'artist_slug', 30)
fig = vislab.dataset_viz.plot_column_frequencies(df, 'artist_slug', 60)
```

### Date distribution plot.
