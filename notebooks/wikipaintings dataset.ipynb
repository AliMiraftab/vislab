# Wikipaintings

This dataset of ~100K paintings and associated genre and style labels (and free-form tags) is scraped from http://wikipaintings.org.

```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import re
import vislab
import vislab.datasets.wikipaintings as wikipaintings
```

Load basic dataset, which is scraped entirely from the search results page and includes the page urls of the artworks.

```python
df = wikipaintings.get_basic_dataset()
df.iloc[:5]
```

Load detailed dataset, which is scraped from the individual artwork pages and includes the labels and image urls of the artworks.

```python
df_d = wikipaintings.get_detailed_dataset()
df_d.iloc[:5]
```

Load the expanded-label dataset, which contains separate binary columns for the top 20 style and genre values.

```python
df_style = wikipaintings.get_style_dataset()
df_style.iloc[0]
```
