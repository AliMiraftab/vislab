```python
label_df = vislab.datasets.flickr.load_flickr_df()
```

```python
c = vislab.util.get_mongodb_client()['predict']['flickr_oct26']
if c.find({'features': 'noise'}).count() > 0:
    c.remove({'features': 'noise'})
pd.DataFrame([x for x in c.find()])
```

```python
df, preds_panel = aphrodite.results.load_pred_results(
    'flickr_oct26', os.path.expanduser('~/work/aphrodite/data/results2'),
    multiclass=True, force=True)
pred_prefix = 'pred'
print preds_panel.minor_axis
```

```python
# Select features to include.

feat_to_evaluate = [
    u'decaf_fc6 False vw',
    u'decaf_fc6_flatten False vw',
    u'decaf_tuned_fc6 False vw',
    #u'decaf_tuned_fc6,pascal_mc_for_decaf_tuned_fc6 pd vw',
    #u'decaf_tuned_fc6_flatten False vw',
    #u'decaf_tuned_fc6_ud None vw',
    #u'fusion_flickr_oct26 None vw',
    u'fusion_flickr_oct26,pascal_mc_for_fusion_flickr_oct26 fp vw',
    u'mc_bit False vw',
    u'decaf_imagenet None vw'
]
preds_panel = preds_panel.select(lambda x: x in feat_to_evaluate, 'minor')
print preds_panel.minor_axis
```

```python
nice_feat_names = {
    'decaf_fc6 False vw': 'DeCAF_6',
    'decaf_fc6_flatten False vw': 'DeCAF_5',
    'decaf_tuned_fc6 False vw': 'Fine-tuned DeCAF_6',
    'decaf_tuned_fc6,pascal_mc_for_decaf_tuned_fc6 pd vw': 'DeCAF_6 x Content',
    'decaf_tuned_fc6_flatten False vw': 'Fine-tuned DeCAF_5',
    'fusion_flickr_oct26 None vw': 'Late fusion',
    'fusion_flickr_oct26,pascal_mc_for_fusion_flickr_oct26 fp vw': 'Late-fusion x Content',
    'mc_bit False vw': 'MC-bit',
    'decaf_imagenet None vw': 'Imagenet'
}

mc_metrics = vislab.results.multiclass_metrics_feat_comparison(
    preds_panel, label_df, pred_prefix, features=preds_panel.minor_axis.tolist() + ['random'],
    balanced=True, with_plot=False, with_print=False, nice_feat_names=nice_feat_names)
```

```python
conf_df = mc_metrics['feat_metrics']['fusion_flickr_oct26,pascal_mc_for_fusion_flickr_oct26 fp vw']['conf_df'].astype(float)
conf_df.index = [x.replace('style_', '') for x in conf_df.index]
conf_df.columns = [x.replace('style_', '') for x in conf_df.columns]
fig = vislab.dataset_viz.plot_conditional_occurrence(conf_df, sort_by_prior=False, font_size=14)
fig.savefig('/Users/sergeyk/work/aphrodite-writeup/figures/evaluation/flickr_conf.pdf', bbox_inches='tight')
```

```python
acc_df = mc_metrics['acc_df']
acc_df = acc_df[['Late-fusion x Content', 'Fine-tuned DeCAF_6', 'DeCAF_6', 'DeCAF_5', 'MC-bit', 'random']]
fig = vislab.results_viz.plot_top_k_accuracies(acc_df, font_size=16)
fig.savefig('/Users/sergeyk/work/aphrodite-writeup/figures/evaluation/flickr_top_k.pdf', bbox_inches='tight')
```

```python
ap_df = mc_metrics['ap_df']
column_order = ap_df.columns[(-ap_df.ix['_mean']).argsort().values]
ap_df.index = [x.replace('style_', '') for x in ap_df.index]
ap_df = ap_df.reindex_axis(column_order, axis=1)
ap_df.to_csv('/Users/sergeyk/work/aphrodite-writeup/results/flickr_ap_df.csv')

fig = vislab.results_viz.plot_df_bar(ap_df, fontsize=14)
fig.savefig('/Users/sergeyk/work/aphrodite-writeup/figures/flickr_ap_barplot.pdf', bbox_inches='tight')
```
